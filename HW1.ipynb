```python
!pwd
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
img = io.imread('/content/drive/MyDrive/AI lab/selfie.jpg')
img.shape
```

1, 사진 사이즈를 줄이고 180도 회전시켜 privacy 보호하기

```python
from skimage.transform import resize
from skimage.transform import rotate
resized_img = resize(img, (img.shape[0]//10, img.shape[1]//10), anti_aliasing=True)
rot180_img = rotate(resized_img, 180, resize=True)
fig, ax = plt.subplots(1, 1)
ax.imshow(rot180_img)
ax.set_title('Resized by 1/10 & 180 degree rotated')
plt.show()
```

2. 얼굴이 나오지 않게 이미지 자르기

```python
crop_img = img[800:1440, 200:800]
io.imshow(crop_img)
```
